###Desafio 07: Deixar os gráficos bonitões. (Matplotlib.pyplot)

```python
import matplotlib.pyplot as plt
dados['tempo'].value_counts().plot(kind='bar', color=['green','blue', 'red'])
plt.xlabel('Tempo')
plt.ylabel('Quantidade de amostras')
plt.title('Quantidades de amostras por tempo de resposta')
plt.show()
```

```python
dados['tratamento'].value_counts().plot(kind='bar', color=['red','blue'])
plt.xlabel('Tipo de tratamento', fontsize = 12)
plt.ylabel('Quantidade de amostras', fontsize = 12)
plt.title('Quantidades de amostras por tipo de tratamento', fontsize=16)
plt.show()
```
